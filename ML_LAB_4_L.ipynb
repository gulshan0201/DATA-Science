# Task 1

## Generate high-dimensional data




```python
import numpy as np
import pandas as pd

# Generate a random dataset with 5000 samples and 1000 features
X = np.random.rand(5000, 1000)

# Display the first few rows of the data
df = pd.DataFrame(X)
display(df.head())
```

## Standardize the data

```python
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

```

PERFORM PCA AND KERNAL PCA

```python
from sklearn.decomposition import PCA, KernelPCA
import pandas as pd

# Perform standard PCA with all components
pca_full = PCA(n_components=None)
X_pca_full = pca_full.fit_transform(X_scaled)

# Reconstruct the data using all components
X_reconstructed_full_pca = pca_full.inverse_transform(X_pca_full)

# Display the first few rows of the reconstructed data from PCA
df_reconstructed_full_pca = pd.DataFrame(X_reconstructed_full_pca)
display(df_reconstructed_full_pca.head())

```

```python
# Perform Kernel PCA with RBF kernel and all components
kpca_full = KernelPCA(n_components=None, kernel='rbf', fit_inverse_transform=True)
X_kpca_full = kpca_full.fit_transform(X_scaled)

# Reconstruct the data using Kernel PCA (Note: reconstruction is not always possible with all kernels)
X_reconstructed_full_kpca = kpca_full.inverse_transform(X_kpca_full)

# Display the first few rows of the reconstructed data from Kernel PCA
df_reconstructed_full_kpca = pd.DataFrame(X_reconstructed_full_kpca)
display(df_reconstructed_full_kpca.head())
```

## Visualize the data

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import pandas as pd

# Perform PCA with 2 components for visualization
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_scaled)

# Create a DataFrame for the 2D PCA results
df_pca_2d = pd.DataFrame(X_pca_2d, columns=['Principal Component 1', 'Principal Component 2'])

# Plot the 2D PCA results
plt.figure(figsize=(8, 6))
plt.scatter(df_pca_2d['Principal Component 1'], df_pca_2d['Principal Component 2'], alpha=0.5)
plt.title('2D PCA Results')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

```python
import plotly.express as px
from sklearn.decomposition import PCA
import pandas as pd

# Perform PCA with 3 components for visualization
pca_3d = PCA(n_components=3)
X_pca_3d = pca_3d.fit_transform(X_scaled)

# Create a DataFrame for the 3D PCA results
df_pca_3d = pd.DataFrame(X_pca_3d, columns=['Principal Component 1', 'Principal Component 2', 'Principal Component 3'])

# Plot the 3D PCA results with modified appearance
fig = px.scatter_3d(df_pca_3d, x='Principal Component 1', y='Principal Component 2', z='Principal Component 3',
                    title='3D PCA Results',
                    size_max=5,  # Adjust marker size
                    opacity=0.7) # Adjust marker opacity
fig.show()
```

# Task 2

```python
n_components_list = [1000, 900, 800, 700, 600, 500, 400, 300, 200, 100]
```

## Iterate and perform pca



```python
import time

reconstructed_data_list = []
pca_times = []

for n_components in n_components_list:
    start_time = time.time()
    pca = PCA(n_components=n_components)
    X_pca = pca.fit_transform(X_scaled)
    X_reconstructed = pca.inverse_transform(X_pca)
    end_time = time.time()
    reconstructed_data_list.append(X_reconstructed)
    pca_times.append(end_time - start_time)
```

## Calculate mean squared error



```python
from sklearn.metrics import mean_squared_error

mse_list = []
for reconstructed_data in reconstructed_data_list:
    mse = mean_squared_error(X_scaled, reconstructed_data)
    mse_list.append(mse)
```

## Store and display results



```python
results_df = pd.DataFrame({
    'Number of Components': n_components_list,
    'Mean Squared Error': mse_list,
    'PCA Time (seconds)': pca_times
})

display(results_df)
```

```python
import matplotlib.pyplot as plt

# Display a scatter plot for each reconstructed dataset
for i, n_components in enumerate(n_components_list):
    plt.figure(figsize=(8, 6))
    plt.scatter(reconstructed_data_list[i][:, 0], reconstructed_data_list[i][:, 1], alpha=0.5)
    plt.title(f'Reconstructed Data with {n_components} Components')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.grid(True)
    plt.show()
```
